```python
def save_obj(obj, name ):
    with open('obj/'+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open('obj/' + name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
import re
import string
from nltk.tokenize import TreebankWordTokenizer
from nltk.stem import PorterStemmer 

ps = PorterStemmer()
alphanumeric = lambda x: re.sub('\w*\d\w*', ' ', x)
punc_lower = lambda x: re.sub('[%s]' % re.escape(string.punctuation), '', x.lower())
punc_n = lambda x: re.sub('\\n', ' ', x)
stem = lambda x: ps.stem(x)
df['Clean_Doc'] = df.Document.map(alphanumeric).map(punc_lower).map(punc_n).map(stem)


df['Clean_doc'] = [[ps.stem(word) for word in doc.split(" ")] for doc in df['Clean_Doc']]
```

```python
import numpy as np
market_cap = []
earning_pershare = []
outstanding_shares = []
pe_ratio = []
div_yield = []

for ticker in df_sectors.Ticker.values:
    market_cap.append(ticker_data[ticker][1][1])
    earning_pershare.append(ticker_data[ticker][5][1])
    outstanding_shares.append(ticker_data[ticker][4][1])
    pe_ratio.append(ticker_data[ticker][6][1])
    div_yield.append(ticker_data[ticker][9][1])
market_cap = np.array(market_cap, dtype = 'U25')
earning_pershare = np.array(earning_pershare, dtype = 'U25')
outstanding_shares = np.array(outstanding_shares, dtype = 'U25')
div_yield = np.array(div_yield, dtype = 'U25')
```

```python
for i, cap in enumerate(market_cap):
    if 'T' in cap:
        market_cap[i] = float(cap[0:-1])*1000000000000
    elif 'B' in cap:
        market_cap[i] = float(cap[0:-1])*1000000000
    elif 'M' in cap:
        market_cap[i] = float(cap[0:-1])*1000000
    elif 'T' in cap:
        market_cap[i] = float(cap[0:-1])*1000
    if ',' in cap:
        market_cap[i] = market_cap[i].replace(',', '')
        
for i, out in enumerate(outstanding_shares):
    if ',' in out:
        outstanding_shares[i] = outstanding_shares[i].replace(',', '')
    if 'T' in out:
        outstanding_shares[i] = round(float(cap[0:-1])*1000000000000)
    elif 'B' in out:
        outstanding_shares[i] = round(float(cap[0:-1])*1000000000)
    elif 'M' in out:
        outstanding_shares[i] = round(float(cap[0:-1])*1000000)
    elif 'T' in out:
        outstanding_shares[i] = round(float(cap[0:-1])*1000)
    
        
for i, div in enumerate(div_yield):
    if '%' in div:
        div_yield[i]= div[0:-1]
    elif 'n/a' in div:
        div_yield[i]= 0
        
for i, eps in enumerate(earning_pershare):
    if 'n/a' in eps:
        earning_pershare[i] = 0
for i, pe in enumerate (pe_ratio):
    if ','in pe:
        pe_ratio[i] = pe_ratio[i].replace(',', '')
    if 'n/a' in pe:
        pe_ratio[i] = 0
    if not pe:
        pe_ratio[i] = 0
    
```

```python
df_sectors['Market Cap'] = market_cap
df_sectors['Outstanding Shares'] = outstanding_shares
df_sectors['EPS'] = earning_pershare
df_sectors['Yield'] = div_yield
df_sectors['PE ratio'] = pe_ratio
```

```python
df_sectors[['Market Cap', 'Outstanding Shares', 'EPS', 'Yield']].fillna(0, inplace = True)
df_sectors[['Market Cap', 'Outstanding Shares', 'EPS', 'Yield', 'PE ratio']] = df_sectors[['Market Cap', 'Outstanding Shares', 'EPS', 'Yield', 'PE ratio']].astype(float)
df_sectors.info()

```

```python
df_sectors['P/E ratio'] = (df_sectors['Market Cap']/df_sectors['Outstanding Shares'])/df_sectors['EPS']
```

```python
df_sectors[df_sectors['PE ratio'] == np.inf]
```

```python
df_sectors['P/E ratio'].replace(np.inf, 0, inplace = True)
df_sectors['PE ratio'].fillna( 0, inplace = True)
df_sectors['P/E ratio'].clip(lower = 0, inplace = True)
```
